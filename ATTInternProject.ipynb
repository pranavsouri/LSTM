```python
data = pd.read_csv("amchart.csv")
```

```python
data['TEST_PERIOD_FORMATTED'] = pd.to_datetime(data['TEST_PERIOD_FORMATTED'])
data.head()
```

```python
data['MA48'] = data['ATT_TP_DL_MV'].rolling(48).mean()
data['MA336'] = data['ATT_TP_DL_MV'].rolling(15).mean()
```

```python
import plotly.express as px
fig = px.line(data, x="TEST_PERIOD_FORMATTED", y=['ATT_TP_DL_MV', 'MA48', 'MA336'], title='Throughput', template = 'plotly_dark')
fig.show()
```

```python
data.set_index('TEST_PERIOD_FORMATTED', drop=True, inplace=True)
```

```python
data['day'] = [i.day for i in data.index]
data['day_name'] = [i.day_name() for i in data.index]
data['day_of_year'] = [i.dayofyear for i in data.index]
data['week_of_year'] = [i.weekofyear for i in data.index]
data['hour'] = [i.hour for i in data.index]
data['is_weekday'] = [i.isoweekday() for i in data.index]
```

```python
iforest = create_model('iforest', fraction = 0.1)
iforest_results = assign_model(iforest)
iforest_results.head()
```

```python
iforest_results[iforest_results['Anomaly'] == 1].head()
```

```python
import plotly.graph_objects as go
fig = px.line(iforest_results, x=iforest_results.index, y="ATT_TP_DL_MV", title='Throughput DL', template = 'plotly_dark')
outlier_dates = iforest_results[iforest_results['Anomaly'] == 1].index
y_values = [iforest_results.loc[i]['ATT_TP_DL_MV'] for i in outlier_dates]
fig.add_trace(go.Scatter(x=outlier_dates, y=y_values, mode = 'markers', 
                name = 'Anomaly', 
                marker=dict(color='red',size=10)))
        
fig.show()
```
